# Merge, Join y Concatenate

En este notebook vamos a ver cómo trabajar y concatenar objetos dentro de Pandas, así como la relación entre los métodos de Pandas y SQL

```python
import pandas as pd
import numpy as np
```

Primero generamos tres DataFrames para trabajar con ellos

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
df1
```

```python
frames = [df1, df2, df3]
frames
```

```python
result = pd.concat(frames, keys=['x', 'y', 'z'])
result
```

```python
result.loc['y']
```

Cuando juntemos distintos DataFrames vamos a tener 3 maneras de hacerlo. La primera es tomar la unión de los DataFrames (join="outer") que es la opción por default dado que no se pierde ninguna información. Segunda, tomar la intersección de los DataFrames (join="inner"). Tercera, utilizar los índices específicos de un dataframe 

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                     'D': ['D2', 'D3', 'D6', 'D7'],
                     'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])
df4
```

```python
df1
```

```python
pd.concat([df1, df4], axis=1)
```

```python
pd.concat([df1, df4], axis=1, join='inner')
```

```python
pd.concat([df1, df4], axis=1, join_axes=[df1.index])
```

### Concatenating using append

```python
df1.append(df2)
```

```python
df1.append(df4)
```

podemos hacer append de múltiples objetos si pasamos una lista de los DataFrames

```python
 df1.append([df2, df3])
```

### Ignorando índices 

Para DataFrames donde tenemos índices que no tienen significado lo mejor es hacerles append e ignorar el índice

```python
pd.concat([df1, df4], ignore_index=True)
```

del mismo modo lo podemos aplicar al método append

```python
df1.append(df4, ignore_index=True)
```

### Concatenando Series a DataFrames

Si tienen nombre, las data series se convierten en una columna cuyo nombre está definido por el nombre de la Series

```python
s1 = pd.Series(['X0', 'X1', 'X2', 'X3'], name='X')
s1
```

```python
pd.concat([df1, s1], axis=1)
```

Si no tienen nombre, las data series se van numerando conforme se agregan al DataFrame

```python
s2 = pd.Series(['_0', '_1', '_2', '_3'])
s2
```

```python
pd.concat([df1, s2, s2, s2], axis=1)
```

### Concatenando con group keys

```python
s3 = pd.Series([0, 1, 2, 3], name='foo')
s4 = pd.Series([0, 1, 2, 3])
s5 = pd.Series([0, 1, 4, 5])
```

```python
pd.concat([s3, s4, s5], axis=1)
```

Vemos que Pandas comienza a numerar a partir de aquellas que no tienen nombre, podemos agregar también manualmente los nombres con el argumento keys

```python
pd.concat([s3, s4, s5], axis=1, keys=['red','blue','yellow'])
```

Lo mismo aplica para los dataframes

```python
pd.concat(frames, keys=['x', 'y', 'z'])
```

o alternativamente

```python
pieces = {'x': df1, 'y': df2, 'z': df3}
pd.concat(pieces)
```

```python
result = pd.concat(pieces, keys=['z', 'y'])
result
```

```python
result.index.levels
```

```python
result = df1.append(s2, ignore_index=True)
result
```

Veamos un caso simple

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                         'A': ['A0', 'A1', 'A2', 'A3'],
                         'B': ['B0', 'B1', 'B2', 'B3']})
left
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
right
```

```python
result = pd.merge(left, right, on='key')
result
```

Aquí vemos un caso más complejo:

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                         'key2': ['K0', 'K1', 'K0', 'K1'],
                         'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
left
```

```python
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                          'key2': ['K0', 'K0', 'K0', 'K0'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
right
```

```python
result = pd.merge(left, right, on=['key1', 'key2'])
result
```

Veamos ejemplos de cada uno de estos casos

```python
# Left Outer Join
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```python
# Right Outer Join
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
# Full Outer Join
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
# Inner Join
pd.merge(left, right, how='inner', on=['key1', 'key2'])
```

## Joins con index

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                         'B': ['B0', 'B1', 'B2']},
                         index=['K0', 'K1', 'K2'])
left
```

```python
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                          'D': ['D0', 'D2', 'D3']},
                          index=['K0', 'K2', 'K3'])
right
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```

```python
left.join(right, how='inner')
```

```python
pd.merge(left, right, left_index=True, right_index=True, how='outer')
```

```python
pd.merge(left, right, left_index=True, right_index=True, how='inner')
```

Es importante notar que las siguientes dos funciones son completamente equivalentes

```python
#left.join(right, on=key_or_keys)
#pd.merge(left, right, left_on=key_or_keys, right_index=True,
#      how='left', sort=False)
```

```python
right = pd.DataFrame({'C': ['C0', 'C1'],
                          'D': ['D0', 'D1']},
                          index=['K0', 'K1'])
right
```

### merge_asof

```python
trades = pd.DataFrame({
        'time': pd.to_datetime(['20160525 13:30:00.023',
                                '20160525 13:30:00.038',
                                '20160525 13:30:00.048',
                                '20160525 13:30:00.048',
                                '20160525 13:30:00.048']),
        'ticker': ['MSFT', 'MSFT',
                   'GOOG', 'GOOG', 'AAPL'],
        'price': [51.95, 51.95,
                  720.77, 720.92, 98.00],
        'quantity': [75, 155,
                     100, 100, 100]},
        columns=['time', 'ticker', 'price', 'quantity'])
```

```python
quotes = pd.DataFrame({
        'time': pd.to_datetime(['20160525 13:30:00.023',
                                '20160525 13:30:00.023',
                                '20160525 13:30:00.030',
                                '20160525 13:30:00.041',
                                '20160525 13:30:00.048',
                                '20160525 13:30:00.049',
                                '20160525 13:30:00.072',
                                '20160525 13:30:00.075']),
        'ticker': ['GOOG', 'MSFT', 'MSFT',
                   'MSFT', 'GOOG', 'AAPL', 'GOOG',
                   'MSFT'],
        'bid': [720.50, 51.95, 51.97, 51.99,
                720.50, 97.99, 720.50, 52.01],
        'ask': [720.93, 51.96, 51.98, 52.00,
                720.93, 98.01, 720.88, 52.03]},
        columns=['time', 'ticker', 'bid', 'ask'])
```

```python
trades
```

```python
quotes
```

merge_asof es parecido a un left-join a excepción de que hace match con la key más cercana en vez de utilizar keys iguales

```python
pd.merge_asof(trades, quotes,
                  on='time',
                  by='ticker')
```

si restringimos la distancia entre el quote y el trade a 2 milisegundos

```python
pd.merge_asof(trades, quotes,
                  on='time',
                  by='ticker',
                  tolerance=pd.Timedelta('2ms'))
```

si restringimos la distancia entre el quote y el trade a 10 milisegundos y quitamos los match exactos en tiempo

```python
pd.merge_asof(trades, quotes,
                  on='time',
                  by='ticker',
                  tolerance=pd.Timedelta('10ms'),
                  allow_exact_matches=False)
```
